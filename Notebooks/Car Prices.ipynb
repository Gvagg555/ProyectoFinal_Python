## Descripción de Columnas 

-   `year` : Columna que muestra el año de producción del auto.
-   `make` : Columna que muestra fábrica del auto.
-   `model`: Columna que muestra modelo del auto.
-   `trim` : Columna que muestra  versión del modelo.
-   `body` : Columna que muestra tipo del vehiculo.
-   `transmision`: Columna que muestra tipo de transmisión.
-   `vin`:  Columna que muestra ID unica del vehiculo (vehicule identification number).
-   `state`: Columna que muestra estado donde se vendió el auto.
-   `condition`: Columna que muestra nota de calidad del estado del vehuculo (-1 a 5+).
-   `odometer`: Columna que muestra millaje del vehiculo.
-   `color`: Columna que muestra color del vehiculo.
-   `interior` : Columna que muestra color del interior del vehiculo.
-   `seller` : Columna que muestra nombre del vendedor del vehivulo.
-   `mmr` : Columna  que muestra el  "Manheim Market Report"  valor establecido para el vehiculo.
-   `sellingprice` : Columna que muestra el valor de venta del vehiculo.
-   `saledate`: Columna que muestra la fecha de venta del vehiculo.

```python
#revision de nulos
null_counts = df.isnull().sum()
null_counts
```

```python
# Corregir columna condition ( valores entre 1 a 5)
def correct_condition(value):
    if value > 5:
        value = value / 10  # Correct the value
    return value

df['condition'] = df['condition'].apply(correct_condition)

condition_corrected = df['condition'].between(1, 5).all()
print("Columna 'Condition' corregida:", condition_corrected)

# revisar los tipos de resultados en esta columna, ya que los NAN los pasaremos a 0
df['condition'] = df['condition'].fillna(0)

unique_values = df['condition'].unique()
print("\nUnique values in the 'condition' column (after filling NaNs with 0):", unique_values)
value_counts = df['condition'].value_counts()
```

```python
#eliminar filas vacias relacionadas al make, ya que  pueden ser autos custom sin marca o errores de la tabla y no los necesitamos para la revisión
df = df.dropna(subset=['make'])
null_counts_after_dropping_make = df.isnull().sum()

#remplazar los NAN por "Not Specified"
df = df.fillna("Not Specified")
null_counts_after_filling = df.isnull().sum()
print("\nNumber of null values in each column after filling NaNs with 'Not Specified':")
print(null_counts_after_filling)

# ver data set
df.head(15)
```
