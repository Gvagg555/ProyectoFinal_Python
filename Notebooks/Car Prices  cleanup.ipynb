## Descripci√≥n de Columnas 

-   `year` :(int64) Columna que muestra el a√±o de producci√≥n del auto.
-   `make` : (object) Columna que muestra f√°brica del auto.
-   `model`: (object) Columna que muestra modelo del auto.
-   `trim` : (object) Columna que muestra  versi√≥n del modelo.
-   `body` : (object) Columna que muestra tipo del vehiculo.
-   `transmision`: (object) Columna que muestra tipo de transmisi√≥n.
-   `vin`:  (object) Columna que muestra ID unica del vehiculo (vehicule identification number).
-   `state`: (object) Columna que muestra estado donde se vendi√≥ el auto.
-   `condition`: (float64) Columna que muestra nota de calidad del estado del vehuculo (-1 a 5+).
-   `odometer`: (float64) Columna que muestra millaje del vehiculo.
-   `color`: (object) Columna que muestra color del vehiculo.
-   `interior` : (object) Columna que muestra color del interior del vehiculo.
-   `seller` : (object) Columna que muestra nombre del vendedor del vehivulo.
-   `mmr` : (object) Columna que muestra el  "Manheim Market Report"  valor establecido para el vehiculo.
-   `sellingprice` : (object) Columna que muestra el valor de venta del vehiculo.
-   `saledate`: (object) Columna que muestra la fecha de venta del vehiculo.

```python
#revision de nulos
null_counts = df.isnull().sum()
null_counts
```

```python
print(df.dtypes)
```

```python
# Corregir columna condition ( valores entre 1 a 5)
def correct_condition(value):
    if value > 5:
        value = value / 10  
    return value

df['condition'] = df['condition'].apply(correct_condition)

condition_corrected = df['condition'].between(1, 5).all()
print("Columna 'Condition' corregida:", condition_corrected)

# revisar los tipos de resultados en esta columna, ya que los NAN los pasaremos a 0
df['condition'] = df['condition'].fillna(0)

unique_values = df['condition'].unique()
print("\nUnique values in the 'condition' column (after filling NaNs with 0):", unique_values)
value_counts = df['condition'].value_counts()
```

```python
#eliminar filas vacias relacionadas al make, ya que  pueden ser autos custom sin marca o errores de la tabla y no los necesitamos para la revisi√≥n
df = df.dropna(subset=['make'])
null_counts_after_dropping_make = df.isnull().sum()

#remplazar los NAN por "Not Specified"
df = df.fillna("Not Specified")
null_counts_after_filling = df.isnull().sum()
print("\nNumber of null values in each column after filling NaNs with 'Not Specified':")
print(null_counts_after_filling)

# ver data set
df.head(10)
```

```python
#modificar la columna saledate
def format_saledate(date_str):
    match = re.match(r'(.*\d{4} \d{2}:\d{2}:\d{2})', date_str)
    if match:
        clean_date_str = match.group(1)
    else:
        # fecha generica para identificar en futuro
        clean_date_str = 'Fri Jul 16 1963 23:59:00'
    
    try:
        dt = datetime.strptime(clean_date_str, '%a %b %d %Y %H:%M:%S')
    except ValueError:
        dt = datetime.strptime('Fri Jul 16 1963 23:59:00', '%a %b %d %Y %H:%M:%S')
    
    return dt

df['formatted_date'] = df['saledate'].apply(format_saledate)

#generar columna para control de ventas por a√±o
df['saleyear'] = df['formatted_date'].dt.year


print(df.head(10))


```
